## Helper Functions

Below are some helper functions to help get you started. They should look familiar from the lesson!

````python
import math

def readimage(image_path):
    image = mpimg.imread(image_path)
    print('This image is:', type(image), 'with dimensions:', image.shape)
    return image

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image

def draw_lines(img, lines, color=[255, 0, 0], thickness=5):
    for line in lines:
        for x1,y1,x2,y2 in line:
            cv2.line(img, (x1, y1), (x2, y2), color, thickness)

            
def fitting_lines(img, lines, color=[255, 0, 0], thickness=10):
    # lines come from 
    
    # classify all lines into left and right
    
    # define initial state
    left_points_x = []
    left_points_y = []
    right_points_x = []
    right_points_y = []
    length_min = 50
    length_lower = 20
    
    global left_points_x_previous
    global left_points_y_previous
    global right_points_x_previous
    global right_points_y_previous
    # seperate left and right slopes based on slope and length of lines
    for line in lines:
        for x1,y1,x2,y2 in line:
            length = np.sqrt((x2 - x1)**2 + (y2 - y1)**2)
            slope = ((y2-y1)/(x2-x1))
            if not math.isnan(slope):
                if length > length_min:
                    if slope < -0.4:
                        left_points_x.append(x1)
                        left_points_x.append(x2)
                        left_points_y.append(y1)
                        left_points_y.append(y2)
                    elif slope > 0.4:
                        right_points_x.append(x1)
                        right_points_x.append(x2)
                        right_points_y.append(y1)
                        right_points_y.append(y2)
    
    print('right: ',len(right_points_x))
    print('left: ',len(left_points_x))
    
    # if either array is too limited, lower length requirement 
    if len(left_points_x)<4 or len(right_points_x)<4:
        for x1,y1,x2,y2 in line:
            length = np.sqrt((x2 - x1)**2 + (y2 - y1)**2)
            slope = ((y2-y1)/(x2-x1))
            if not math.isnan(slope):
                if length > length_lower:
                    if slope < -0.35:
                        left_points_x.append(x1)
                        left_points_x.append(x2)
                        left_points_y.append(y1)
                        left_points_y.append(y2)
                    elif slope > 0.35:
                        right_points_x.append(x1)
                        right_points_x.append(x2)
                        right_points_y.append(y1)
                        right_points_y.append(y2)
    
        print('right_add: ',len(right_points_x))
        print('left_add: ',len(left_points_x))
    
    # if either array is empty, keep last valid one
    # record last valid one
    if not len(left_points_x)==0:
        left_points_x_previous = left_points_x
        left_points_y_previous = left_points_y    
        print('left previous: ', left_points_x_previous)
    
    if not len(right_points_x)==0:
        right_points_x_previous = right_points_x
        right_points_y_previous = right_points_y
        print('right previous: ', right_points_x_previous)
    
    # using last one 
    if len(left_points_x)==0:
        left_points_x = left_points_x_previous
        left_points_y = left_points_y_previous
    
    if len(right_points_x)==0:  
        right_points_x = right_points_x_previous
        right_points_y = right_points_y_previous
    
    # make an array for all selected points
    left_points_x = np.array(left_points_x)
    left_points_y = np.array(left_points_y)
    right_points_x = np.array(right_points_x)
    right_points_y = np.array(right_points_y)
                        
    # using 1-dimension polyfit to calculate only slope and 
    k_left,b_left = np.polyfit(left_points_x,left_points_y,1)
    k_right,b_right = np.polyfit(right_points_x,right_points_y,1)
    #print("left line: ", k_left,b_left)
    #print("right line: ",k_right,b_right)
    
    # get dominant line points using x = (y-b)/k
    imshape = img.shape
    ysize = img.shape[0]
    xsize = img.shape[1]
    x_proporation_1 = 0.45
    x_proporation_2 = 0.55
    y_proporation = 0.7
    y_upper = y_proporation * ysize
    y_lower = ysize
    left_point1 = (int((y_upper-b_left)/k_left),int(y_upper))
    left_point2 = (int((y_lower-b_left)/k_left),int(y_lower))
    right_point1 = (int((y_upper-b_right)/k_right),int(y_upper))
    right_point2 = (int((y_lower-b_right)/k_right),int(y_lower))
    #print('left points: ',left_point1,left_point2)
    #print('right points: ',right_point1,right_point2)
    
    # draw lines
    cv2.line(img, left_point1, left_point2, color, thickness)
    cv2.line(img, right_point1, right_point2, color, thickness)   

def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    #draw_lines(line_img, lines)
    #fitting_lines(line_img, lines)
    return line_img

# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, α=0.8, β=1., λ=0.):

  
    return cv2.addWeighted(initial_img, α, img, β, λ)

def image2lane_lines(image_path):
    # reading in an image
    image = mpimg.imread(image_path)
    #printing out some stats and plotting
    print('This image is:', type(image), 'with dimensions:', image.shape)
    #plt.imshow(image)
    # convert image to grayscale
    gray = grayscale(image)
    
    # Gaussian smooth
    blur_gray = gaussian_blur(gray, 5)
    
    # Canny edge detection
    edges_image = canny(blur_gray, 30, 100)
    
    # Mask edges image
    imshape = image.shape
    ysize = image.shape[0]
    xsize = image.shape[1]
    x_proporation_1 = 0.44
    x_proporation_2 = 0.56
    y_proporation = 0.6
    vertices = np.array([[(0,ysize) ,(x_proporation_1*xsize,y_proporation*ysize),(x_proporation_2*xsize,y_proporation*ysize), (xsize ,ysize)]], dtype=np.int32)
    edges_image_with_mask = region_of_interest(edges_image, vertices)
    
    #  Hough lines
    rho = 1 # distance resolution in pixels of the Hough grid
    theta = np.pi/180 # angular resolution in radians of the Hough grid
    threshold = 10     # minimum number of votes (intersections in Hough grid cell)
    min_line_length = 30 #minimum number of pixels making up a line
    max_line_gap = 20
    lines = cv2.HoughLinesP(edges_image_with_mask, rho, theta, threshold, np.array([]), min_line_length, max_line_gap)
    lines_image1 = np.zeros((image.shape[0], image.shape[1], 3), dtype=np.uint8)
    lines_image2 = np.zeros((image.shape[0], image.shape[1], 3), dtype=np.uint8)
    #lines_image = hough_lines(edges_image_with_mask, rho, theta, threshold, min_line_length, max_line_gap)
    # draw hough lines
    plt.subplot(1,2,1)
    draw_lines(lines_image1, lines)
    hough_image = weighted_img(lines_image1, image)
    plt.imshow(hough_image)
    fitting_lines(lines_image2, lines)
    # Combine lines image with original image
    final_image = weighted_img(lines_image2, image)
    ## Plot final image
    plt.subplot(1,2,2)
    plt.imshow(final_image)

    return final_image

#iamge = readimage('test_images/WhiteCarLaneSwitch.jpg')
image2lane_lines('test_images/WhiteCarLaneSwitch.jpg')


def video2lane_lines(image):
    # reading in an image
    #image = mpimg.imread(image_path)
    #printing out some stats and plotting
    print('This image is:', type(image), 'with dimensions:', image.shape)
    #plt.imshow(image)
    # convert image to grayscale
    gray = grayscale(image)
    
    # Gaussian smooth
    blur_gray = gaussian_blur(gray, 5)
    
    # Canny edge detection
    edges_image = canny(blur_gray, 30, 100)
    
    # Mask edges image
    imshape = image.shape
    ysize = image.shape[0]
    xsize = image.shape[1]
    x_proporation_1 = 0.42
    x_proporation_2 = 0.58
    y_proporation = 0.6
    vertices = np.array([[(0,ysize) ,(x_proporation_1*xsize,y_proporation*ysize),(x_proporation_2*xsize,y_proporation*ysize), (xsize ,ysize)]], dtype=np.int32)
    edges_image_with_mask = region_of_interest(edges_image, vertices)
    
    #  Hough lines
    rho = 1 # distance resolution in pixels of the Hough grid
    theta = np.pi/180 # angular resolution in radians of the Hough grid
    threshold = 10     # minimum number of votes (intersections in Hough grid cell)
    min_line_length = 30 #minimum number of pixels making up a line
    max_line_gap = 15
    lines = cv2.HoughLinesP(edges_image_with_mask, rho, theta, threshold, np.array([]), min_line_length, max_line_gap)
    lines_image = np.zeros((image.shape[0], image.shape[1], 3), dtype=np.uint8)    
    fitting_lines(lines_image, lines)
    
    # Combine lines image with original image
    final_image = weighted_img(lines_image, image)
    
    # Plot final image
    #plt.imshow(final_image)
    return final_image

def video2lane_lines_Challenge(image):
    # reading in an image
    #image = mpimg.imread(image_path)
    #printing out some stats and plotting
    print('This image is:', type(image), 'with dimensions:', image.shape)
    #plt.imshow(image)
    # convert image to grayscale
    gray = grayscale(image)
    
    # Gaussian smooth
    blur_gray = gaussian_blur(gray, 3)
    
    # Canny edge detection
    edges_image = canny(blur_gray, 30, 80)
    edges_image = canny(blur_gray, 50, 150)
    #edges_image = canny(image, 25, 55)
    # Mask edges image
    imshape = image.shape
    ysize = image.shape[0]
    xsize = image.shape[1]
    x_proporation_1 = 0.45
    x_proporation_2 = 0.55
    y_proporation_1 = 0.58
    y_prop_2 = 0.9
    vertices = np.array([[(0.25*xsize,ysize*y_prop_2) ,(x_proporation_1*xsize,y_proporation*ysize),(x_proporation_2*xsize,y_proporation*ysize), (0.75*xsize ,ysize*y_prop_2)]], dtype=np.int32)
    edges_image_with_mask = region_of_interest(edges_image, vertices)
    
    #  Hough lines
    rho = 2 # distance resolution in pixels of the Hough grid
    theta = np.pi/180 # angular resolution in radians of the Hough grid
    threshold = 5     # minimum number of votes (intersections in Hough grid cell)
    min_line_length = 10 #minimum number of pixels making up a line
    max_line_gap = 20
    lines = cv2.HoughLinesP(edges_image_with_mask, rho, theta, threshold, np.array([]), min_line_length, max_line_gap)
    lines_image = np.zeros((image.shape[0], image.shape[1], 3), dtype=np.uint8)    
    fitting_lines(lines_image, lines)
    
    # Combine lines image with original image
    final_image = weighted_img(lines_image, image)
    # Plot final image
    #plt.imshow(final_image)
    return final_image
````
